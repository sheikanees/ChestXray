Pneumonia is defined by the Oxford Dictionary as lung inflammation caused by bacterial or viral infection, in which the air sacs fill with pus and may become solid.  Pneumonia, one of the leading causes of deaths in Singapore  accounted for 18.5 per cent of deaths in 2013.

Causes for Pneumonia include Flu viruses, cold viruses, RSV virus, Bacteria called Streptococcus pneumoniae and Mycoplasma pneumoniae.

An x-ray exam of the chest will allow the doctor to see the lungs, heart and blood vessels to help determine if one has pneumonia. When interpreting the x-ray, the radiologist will look for white spots in the lungs (called infiltrates) that identify an infection. 

A machine learning algorithm that can automate the checking for pneumonia based on Chest X-ray images can result in much faster processing and hence faster treatment for patients.  

The dataset was obtained from kaggle:
https://www.kaggle.com/paultimothymooney/chest-xray-pneumonia

Contains: 5216 training photos, 624 test photos and 16 validation photos.

### Forming the training, test and validation dataframes from the raw files that can be fed into tf.data

```python
def loading_df(dataset_string):
    path = dataset_string
    data = []
    for root, dirs, files in os.walk(path, topdown=True):
      for name in files:
        if '.jpeg' not in name:
          continue
        filename = os.path.abspath(os.path.join(root, name))
        class_name = os.path.basename(root)
        data.append((filename, class_name))

    df = pd.DataFrame(data, columns=['filename', 'class_name'])
    df['class_name'] = df['class_name'].astype('category')
    df['class'] = df['class_name'].cat.codes

    #shuffle
    df = df.sample(frac=1).reset_index(drop=False)

    print(df.head())
    print("df shape = {}".format(df.shape))
    print(df.dtypes)
    print(df['class_name'].value_counts())
    
    return df

#Loading training dataframe:
train_df = loading_df("train")

#Loading test dataframe:
test_df = loading_df("test")

#Loading validation dataframe:
val_df = loading_df("val")
```

### Pipeline for loading and resizing images and data augmentation

```python
# Reads an image from a file, decodes it into a tensor, performs image augmentation and resizes it
# to a fixed shape.
img_rows, img_cols = 224,224
num_classes = 2
batch_size = 32

def _parse_function(filename, label):
  image_string = tf.io.read_file(filename)
  image_decoded = tf.image.decode_jpeg(image_string,channels=3)
    
  # Image augmentation 
  image_augmented = tf.image.random_flip_left_right(image_decoded,seed=23)
  # Contrast and Brightness
  #image_augmented = tf.image.random_contrast( image_augmented, 0.1,0.9,seed=None)
  #image_augmented = tf.image.random_brightness( image_augmented, max_delta= 0.05, seed=None)
  #image_augmented = tf.image.random_hue( image_augmented, max_delta= 0.05, seed=None)
    
  image_resized = tf.image.resize(image_augmented, [img_rows, img_cols])
  image_resized = tf.ensure_shape(image_resized ,shape=(img_rows, img_cols,3))
  image_processed = tf.keras.applications.inception_v3.preprocess_input(image_resized) 
  label = tf.one_hot(label, num_classes)
  return image_processed, label
```

### Forming the loading, test and validation datasets using tf.data

```python
train_dataset = tf.data.Dataset.from_tensor_slices((tf.cast(train_df['filename'].values, tf.string),
                                                    tf.cast(train_df['class'].values, tf.int32) ))
train_dataset = train_dataset.map(_parse_function)
train_dataset = train_dataset.shuffle(8000)
train_dataset = train_dataset.repeat()
train_dataset = train_dataset.batch(batch_size, drop_remainder=True)
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices((tf.cast(test_df['filename'].values, tf.string),
                                                    tf.cast(test_df['class'].values, tf.int32) ))
test_dataset = test_dataset.map(_parse_function)
test_dataset = test_dataset.shuffle(8000)
test_dataset = test_dataset.repeat()
test_dataset = test_dataset.batch(batch_size, drop_remainder=True)
```

```python
valid_dataset = tf.data.Dataset.from_tensor_slices((tf.cast(val_df['filename'].values, tf.string),
                                                    tf.cast(val_df['class'].values, tf.int32) ))
valid_dataset = valid_dataset.map(_parse_function)
valid_dataset = valid_dataset.repeat()
valid_dataset = valid_dataset.batch(batch_size, drop_remainder=True)
```

### Transfer learning with the VGG16 model and weights pretrained by ImageNet

```python
base_model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3),pooling=None)
```

```python
# adding a global spatial average pooling layer
x = base_model.output
x = Dropout(0.3)(x)
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)

# adding 2 fully-connected layers 
x = Dense(1024, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.3)(x)

#x = Flatten()(x)

# adding a logits layer (for 2 classes => Normal and Pneumonia)
predictions = Dense(2, activation='softmax')(x)

# this is the model we will train
model = Model(inputs=base_model.input, outputs=predictions)
```

```python
# Transfer learning => training only non base_model layers, i.e. freeze all base_model layers
for layer in base_model.layers:
    #if 'batch_normalization' in layer.name:
     # layer.trainable = True
    #else:
    layer.trainable = False
    
for layer in model.layers:
    print(layer.name,' Trainable =',layer.trainable)
```

```python
model.summary()
```

### Creating the model:

```python
opt = tf.keras.optimizers.Adam(lr=0.001)

# compiling the model
model.compile(optimizer=opt, loss='categorical_crossentropy',metrics=['accuracy'])

# checkpoint
checkpoint_path = "train/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

# Create checkpoint callback
cp_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                 save_weights_only=True, save_best_only=True,
                                                 verbose=1)
```

### Training the model

```python
train_steps = int(5216/32) #total trains set / batch_size
test_steps = int(624/32) #total test set/batch_size
epochs = 1

### Train the model with test set 
history = model.fit( train_dataset, 
                    steps_per_epoch = train_steps,
                   epochs = epochs,
                   validation_data = test_dataset,
                   validation_steps = test_steps, callbacks = [cp_callback])
```

### Evaluating the model

```python
val_steps = 1

metrics = model.evaluate(valid_dataset,
                   steps = val_steps)
print("model accuracy:",metrics[1])
```
